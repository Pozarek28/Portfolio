#### Numerical Methods - Lagrange Interpolation

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Initialization of lists X and Y, you can enter any number of points
# Example data from the presentation
X = np.array([-1, 1, 3, 4, 5, 7, 11, -2])
Y = np.array([4, 0, 4, 9, 16, 20, 5, 9])

# Display the entered points
print("Entered points:")
for i, (x,y) in enumerate(zip(X,Y), start=1):
    print(f"Point {i}: X{i} = {x}, Y{i} = {y}")

# Function calculating the values of the Lagrange interpolation polynomial
def lagrange_interpolation(xi):
    result = 0
    n = len(X)
    for i in range(n):
        value = Y[i]
        for j in range(n):
            if j != i:
                value = value * (xi - X[j]) / (X[i] - X[j])
        result =  result + value
    return result

# Lagrange interpolation polynomial
lagrange_polynomial = lagrange_interpolation

# Prepare data for the plot
x_range = np.linspace(min(X), max(X), 100)
y_range = [lagrange_polynomial(xi) for xi in x_range]

# Plotting the graph
plt.scatter(X, Y, color='red', label='Selected coordinates')
plt.plot(x_range, y_range, label='Lagrange Polynomial', linestyle='dashed')
plt.title('Lagrange Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
